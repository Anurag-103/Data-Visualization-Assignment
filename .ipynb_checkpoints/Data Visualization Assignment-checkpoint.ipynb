## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),linestyle='dashdot',color='k')
plt.xlabel('X',size=20)
plt.ylabel("Y",size=20)
plt.title('Sine Wave Graph',size=20)
```

```python
l=[2,3,4,5]
plt.plot(l,np.sin(l),"b--",marker='H')
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
States=['Delhi',"Mumbai",'Channai',"Kolkata",'Hyderabad']
clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
var=[x for x,_ in enumerate(States)]
plt.bar(var,clothing,width=0.8,label="clothing",color='gold',bottom=Equipment+Accessories)
plt.bar(var,Equipment,width=0.8,label="Equipment",color='silver',bottom=Accessories)
plt.bar(var,Accessories,width=0.8,label="Accessories",color='brown')
plt.ylabel('Items',size=20)
plt.xlabel("States",size=20)
plt.xticks(var,States,rotation=90)
plt.title("RETAILER DATABASE",size=20)
plt.grid(False)
plt.legend()
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.barplot(x='class',y='survived',hue='class',palette='Accent_r',data=titanic)
plt.legend(loc='upper left')
plt.show()
```

## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
d2 = pd.DataFrame({'Country':['United States','Great Britain','India','Russia','Germany'],
     'Medal':[46,27,26,19,17]})
d2
exp = (0,0,0.1,0,0)
colors=['green','red','blue','orange','lightblue']
plt.pie(d2.Medal,labels=d2.Country,colors=colors,autopct='%.2f%%',explode=exp,shadow=1)
plt.show()
```
